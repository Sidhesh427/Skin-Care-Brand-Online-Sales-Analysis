```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading datasets and basic sanitation
channel_data=pd.read_csv('assignment_data_channel.csv')
channel_data.columns=channel_data.columns.str.lower()
channel_data.head()
```

```python
website_data=pd.read_csv('assignment_data_website.csv')
website_data.columns=website_data.columns.str.lower()
website_data.head()
```

```python
sales=pd.read_csv('assignment_data_sales.csv')
sales.columns=sales.columns.str.lower()
sales.rename(columns={'total quatity sold':'total quantity sold'},inplace=True)
sales.head()
```

```python
# creating a caopy of data to use for EDA
sales_copy=sales.copy()
sales_copy.head()
```

```python
# adding additional required rows for analysis
sales_copy['promo_qty_sold']=sales_copy['total quantity sold']-sales_copy['regular quantity sold']
sales_copy['promo_sales']=sales_copy['promo_qty_sold']*27.5
sales_copy['reg_sales']=sales_copy['total $ sales']-sales_copy['promo_sales']
sales_copy
```

```python
# grouping data based on channel to calculate sum across channel
chnl_grp_sales=sales_copy.groupby('channel',as_index=False)['total $ sales','promo_sales','reg_sales'].sum()
pd.DataFrame(chnl_grp_sales)
```

```python
# visualising the data
%matplotlib inline
from matplotlib.ticker import FuncFormatter
def millions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fM' % (x * 1e-6)
formatter = FuncFormatter(millions)

fig,ax= plt.subplots(figsize=(10,7))
fig.patch.set_facecolor('white')
plt.style.use('fivethirtyeight')
plt.rcParams['text.color'] = '#000000'
plt.rcParams['axes.labelcolor']= '#000000'
plt.rcParams['xtick.color'] = '#000000'
plt.rcParams['ytick.color'] = '#000000'
plt.rcParams['font.size']=14
colors=['#80BD9E','#F98866','#336B87','#A1D6E2']
width=0.4
pos=np.arange(len(chnl_grp_sales['channel']))
bar1=ax.bar(pos,chnl_grp_sales['total $ sales'],color=colors,width=width)
ax.set_xticks(pos)
ax.set_xticklabels(list(chnl_grp_sales['channel']))
ax.yaxis.set_major_formatter(formatter)
ax.set_title('CHANNEL SALES',fontsize=17)
plt.savefig('channelsales.png')
plt.show()
```

```python
# joing sales data to channel and website data
id_join=pd.merge(sales_copy,channel_data,on=['account_id','channel'])
id_join
df=pd.merge(id_join,website_data,on='id')
df
```

```python
# grouping dataset by website for sales figures
dataset=df.groupby('website',as_index=False)['promo_sales','reg_sales','total $ sales'].sum()
dataset
```

```python
# visualising sales across websites
from matplotlib.ticker import FuncFormatter
def millions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fM' % (x * 1e-6)
formatter = FuncFormatter(millions)

websites=['Amazon','Flipkart','Snapdeal']
fig,ax= plt.subplots(figsize=(10,7))
fig.patch.set_facecolor('white')
plt.style.use('fivethirtyeight')
plt.rcParams['text.color'] = '#000000'
plt.rcParams['axes.labelcolor']= '#000000'
plt.rcParams['xtick.color'] = '#000000'
plt.rcParams['ytick.color'] = '#000000'
plt.rcParams['font.size']=14
colors=['#FFD64D','#31A9B8','#EB5E30']
width=0.4
pos=np.arange(len(dataset['website']))
bar1=ax.bar(pos,dataset['total $ sales'],color=colors,width=width)
ax.set_xticks(pos)
ax.set_xticklabels(websites)
ax.yaxis.set_major_formatter(formatter)
ax.set_title('WEBSITE SALES',fontsize=17)
plt.savefig('websitesales.png')
plt.show()
```

```python
# visualising sales split across websites in promo and non-promo sales
websites=['Amazon','Flipkart','Snapdeal']
pos=np.arange(len(websites))
promo_sales_total=list(dataset['promo_sales'])
reg_sales_total=list(dataset['reg_sales'])
bar_width=0.30
lables=['promo_sales_total','reg_sales_total']
fig, ax=plt.subplots(figsize=(10,7))
fig.patch.set_facecolor('white')
plt.style.use('fivethirtyeight')
bar1=ax.bar(pos-bar_width/2,promo_sales_total, bar_width,label='promo_sales_total',color='#9A9EAB')
bar2=ax.bar(pos+bar_width/2,reg_sales_total, bar_width, label='reg_sales_total',color='#EC96A4')
ax.set_xticks(pos)
ax.set_xticklabels(websites)
ax.yaxis.set_major_formatter(formatter)
ax.set_title('PROMO vs NON-PROMO SALES ACROSS WEBSITES',fontsize=20)
ax.legend(fontsize=18,ncol=2)
plt.savefig('SalesAcrossWebsites.png')
plt.show()
```

```python
# grouping dataset by channel to get promo vs non-promo sales
dataset_channel=df.groupby('channel',as_index=False)['promo_sales','reg_sales'].sum()
dataset_channel
```

```python
# visualizing sales split across websites in promo and non-promo sales
channels=list(dataset_channel['channel'])
pos=np.arange(len(channels))
promo_sales_total=list(dataset_channel['promo_sales'])
reg_sales_total=list(dataset_channel['reg_sales'])
bar_width=0.30
fig, ax=plt.subplots(figsize=(10,7))
fig.patch.set_facecolor('white')
plt.style.use('fivethirtyeight')
bar1=ax.bar(pos-bar_width/2,promo_sales_total, bar_width, label='promo_sales_total',color='#5D535E')
bar2=ax.bar(pos+bar_width/2,reg_sales_total, bar_width, label='reg_sales_total',color='#DFE166')
ax.set_xticks(pos)
ax.set_xticklabels(channels)
ax.yaxis.set_major_formatter(formatter)
ax.set_title('PROMO vs NON-PROMO SALES ACROSS CHANNELS')
ax.legend(fontsize=17,ncol=2)
plt.savefig('SalesAcrossChannels.png')
plt.show()
```

```python
# adding % columns to the datases
dataset['promo_sales_perct']=round((dataset['promo_sales']/dataset['promo_sales'].sum())*100,0)
dataset['reg_sales_perct']=round((dataset['reg_sales']/dataset['reg_sales'].sum())*100,0)
dataset
```

```python
# Pie chart for % sales for promo vs non-promo across websites
fig=plt.figure()
fig.patch.set_facecolor('white')
labels=['Amazon','Flipkart','Snapdeal']
colors=['#FFDB5C','#4897D8','#FA6E59']
explode=[0.2,0,0]
axes1=fig.add_axes([0.1,0.1,0.5,0.5])
axes1.pie(dataset['promo_sales_perct'], radius=2.5, explode=explode, shadow=True, 
          colors=colors, autopct='%1.1f%%', textprops={'fontsize':15,'fontweight':'bold'})
axes1.set_title('PROMO SALES', loc='center', y=1.8, fontsize=17, fontweight='bold')

axes2=fig.add_axes([1,0.1,0.5,0.5])
axes2.pie(dataset['reg_sales_perct'], radius=2.5, explode=explode, shadow=True, 
          colors=colors, autopct='%1.1f%%',textprops={'fontsize':15,'fontweight':'bold'})
axes2.set_title('REGULAR SALES', loc='center', y=1.8,fontsize=17, fontweight='bold')
axes1.legend(labels,title='Websites', loc='upper right', bbox_to_anchor=(2.1,-1.7,1.0,1.0),
             ncol=3, fontsize=14)
plt.savefig('SalesDistPieW.png',bbox_inches='tight')
plt.show()


```
